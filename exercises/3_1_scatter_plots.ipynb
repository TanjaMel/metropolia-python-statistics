```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the scatter plots
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.rand(100) * 100

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
cmaps = ['seismic','CMRmap','viridis','plasma']
# Customize each subplot
for i, ax in enumerate(axs.flat):
    scatter = ax.scatter(x, y, s=sizes, c=colors, cmap=cmaps[i], alpha=0.2+i*0.2, edgecolors='black', linewidths=1)
    ax.set_title(f'Scatter Plot {i+1}')
    ax.grid(True)

# Set common labels for the entire figure
fig.suptitle('Scatter Plots', fontsize=20)
fig.supxlabel('X-axis')
fig.supylabel('Y-axis')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load iris data from the publicly available UCI archive
file_name = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(file_name)

# Get column names
cols = df.columns
plt.figure(figsize=(10, 8))

# Group the data by species and plot each group
for name, group in df.groupby('species'):     # Using groupby function to seperate data by species to plot them individually
    plt.scatter(
        group[cols[0]], group[cols[1]],
        alpha=0.5,
        edgecolors='black',
        label=name  # Add a label for each species group
    )

plt.title(f"{cols[0]} vs {cols[1]}")
plt.xlabel(f"{cols[0]} cm")
plt.ylabel(f"{cols[1]} cm")
# The legend automatically works as we have plotted already by species seperately
plt.legend(title="Species")
plt.show()
```
