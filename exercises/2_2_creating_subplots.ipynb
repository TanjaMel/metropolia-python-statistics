```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/10)
y4 = np.exp(x)
df = pd.DataFrame({'x':x,'y':y4})
# Create a figure and a grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot on each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Exponential Decay')

axes[1, 1].plot(df['x'], df['y'])
axes[1, 1].set_title('Exponential curve')

# Labeling axes of plots
for i in range(2):
    for j in range(2):
        axes[i, j].set_xlabel('x axis')
        axes[i, j].set_ylabel('y axis')

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
