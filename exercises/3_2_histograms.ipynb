```python
import pandas as pd
import matplotlib.pyplot as plt

# Load iris data
file_name = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(file_name)

# Create a figure and axis objects
fig, axes = plt.subplots(2, 2, figsize=(15,9))
fig.suptitle("Iris Dataset Histograms", fontsize=18)     # Title for whole subplot

# Basic Histogram with Default Settings
axes[0, 0].hist(df['sepal_length'])
axes[0, 0].set_title("Sepal Length Distribution")
axes[0, 0].set_xlabel("Sepal Length")
axes[0, 0].set_ylabel("Frequency")

# Histogram with Custom Bins and Color
axes[0, 1].hist(df['sepal_width'], bins=20, color='purple')
axes[0, 1].set_title("Sepal Width Distribution")
axes[0, 1].set_xlabel("Sepal Width")
axes[0, 1].set_ylabel("Frequency")

# Horizontal histogram with custon bins and color
axes[1, 0].hist(df['sepal_length'], bins=10, color='blue', orientation='horizontal')
axes[1, 0].set_title("Sepal Length Distribution - Horizontal Histogram")
axes[1, 0].set_xlabel("Frequency")
axes[1, 0].set_ylabel("Sepal Length")

# Stacked Histogram by Species
species = df['species'].unique()
colors = ['red', 'blue', 'green']

axes[1, 1].hist(
    [df[df['species'] == s]['sepal_length'] for s in species],
    bins=15,
    color=colors,
    label=species,
    histtype='barstacked'
)

axes[1, 1].set_title("Sepal Length Distribution by Species")
axes[1, 1].set_xlabel("Sepal Length")
axes[1, 1].set_ylabel("Frequency")
axes[1, 1].legend(title="Species")

plt.tight_layout()
plt.show()
```
