```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate linear data
np.random.seed(2)
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + np.random.normal(0, 1, 100)       # Linear relationship

# Generate non-linear data
x_nonlinear = np.linspace(0, 10, 100)
y_nonlinear = x_nonlinear**2 + np.random.normal(0, 1, 100)  # Quadratic relationship

# Convert to DataFrame for easier handling with Seaborn
df_linear = pd.DataFrame({'x': x_linear, 'y': y_linear})
df_nonlinear = pd.DataFrame({'x': x_nonlinear, 'y': y_nonlinear})

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Linear Data - Residual plot
sns.residplot(x='x', y='y', data=df_linear, order=1, ax=axes[0])
axes[0].set_title('Residual Plot for Linear Data')

# Non-linear Data - Residual plot
sns.residplot(x='x', y='y', data=df_nonlinear, order=1, ax=axes[1])
axes[1].set_title('Residual Plot for Non-linear Data')
plt.show()
```
