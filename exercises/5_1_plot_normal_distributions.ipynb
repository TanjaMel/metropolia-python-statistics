Create a 2 by 2 subplot. Plot 4 gaussian distributions, with two different means (0, 10) and two standard deviations (0, 4). Have the rows have the same mean and columns same standard deviation. Have axises labeled with x and y and give each plot a legend that describes the curve. Give the whole figure a title "Gaussian distributions with varying parameters". Have the color of the line in the plots be red. Have your x values be linearly spaced between -10 and 20 for plotting. Copy paste your code in the answer box and attach the outputted image.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 20, 500)

means = [0, 10]
stds = [0, 4]


fig, axes = plt.subplots(2, 2, figsize=(8,6))

for i, mean in enumerate(means):
    for j, std in enumerate(stds):

        # gaussian formula
        if std == 0:
            y = np.zeros_like(x)
        else:
            y = (1/(std*np.sqrt(2*np.pi))) * np.exp(-((x-mean)**2)/(2*std**2))

        axes[i,j].plot(x, y, color='red', label=f"mean={mean}, std={std}")

        axes[i,j].set_xlabel("x")
        axes[i,j].set_ylabel("y")
        axes[i,j].legend()


fig.suptitle("Gaussian distributions with varying parameters")

plt.tight_layout()
plt.show()
```
