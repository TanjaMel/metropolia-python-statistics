Your task is to sample 10, 100, 1000, 10000 datapoints from the standard normal distribution. Then plot the distributions of your samples on 4 subplots using historgrams. Have the binsize set to 31. Label the axes with "Value" for x and "Frequency" for y and give the figure the title "Sampling from the Standard Normal Distribution". Have the color be blue and the edgecolor be black. Give each subplot a title giving information about the sample size. Copy paste your code in the answer box and attach the outputted image.

```python
import numpy as np
import matplotlib.pyplot as plt

sizes = [10, 100, 1000, 10000]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, n in enumerate(sizes):

    data = np.random.normal(0, 1, n)

    axes[i].hist(data, bins=31, color="blue", edgecolor="black")

    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

    axes[i].set_title(f"Sample size = {n}")

fig.suptitle("Sampling from the Standard Normal Distribution")

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```
