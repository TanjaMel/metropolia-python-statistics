```python
import pandas as pd

file = "data/quarterly_financials.csv"   # Можешь подставить свой путь
df = pd.read_csv(file)

# Filter only McDonald's (ticker MCD)
df = df[df["stock"] == "MCD"]

# Set index to filing_date
df = df.set_index("filing_date")

df = df.sort_index()

# Drop rows with ANY NaN values
df = df.dropna()

# Print cleaned head
print(df.head())

# Harmonic mean of earningspershare
hmean = len(df["earningspershare"]) / (1 / df["earningspershare"]).sum()

print(f"\nThe harmonic mean of P/E values: {hmean}")
```
