```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,1)          # 50 evenly spaced numbers from 0 to 10
y = 2*x + 3                   # Creating array y so each element is respective element in x multiplied by 2 and added 3 afterS
```

```python
plt.plot(x,y,"ob", label="Blue dots")                # Scatter plot of first points, (x,y) as blue circles
```

```python
y2 = 2*y
plt.plot(x, y2, label="Line")                        # Plotting a line of connecting points (x, y2)  
```

```python
plt.title("My graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```
