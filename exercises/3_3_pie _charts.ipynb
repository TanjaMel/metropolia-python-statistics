```python
import pandas as pd
import matplotlib.pyplot as plt

# Load iris data
file_name = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(file_name)

# Group data by species to use for pie charts
species_counts = df['species'].value_counts()     # Series with the value counts of unique values as the data and index
                                                  # is the value, in this case our 3 species and 50 each

# Create a figure and axis objects
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle("Iris Dataset Pie Charts", fontsize=16)

# 1. Basic Pie Chart
axes[0].pie(species_counts, labels=species_counts.index)
axes[0].set_title("Species Distribution")

# 2. Pie Chart with Autopct and Explode
explode = [0.1, 0, 0.1]  # Explode the first and last slices
colors = ['gold', 'yellowgreen', 'lightcoral']

axes[1].pie(species_counts, labels=species_counts.index, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True)
axes[1].set_title("Species Distribution with Percentages & Explode")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
